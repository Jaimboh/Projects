The percentage of people who do not have a job but are actively looking for one is called the unemployment rate. During the Covid-19 pandemic, we have seen a significant increase in this rate. As a result, studying the unemployment rate could be a valuable data science project.

## Unemployment Analysis 

To examine unemployment, I will be using an unemployment dataset specifically for India. This dataset includes information on unemployment in India during the Covid-19 pandemic. To begin this analysis, I will need to import the required Python libraries and the dataset

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/unemployment.csv")
print(data.head())
```

Let’s see if this dataset contains missing values or not:

```python
print(data.isnull().sum())
```

During the analysis of missing values, I discovered that the column names are incorrect. To improve our understanding of the data, I will be renaming all of the columns.

```python
data.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate",
               "Estimated Employed",
               "Estimated Labour Participation Rate",
               "Region","longitude","latitude"]
```

Now let’s have a look at the correlation between the features of this dataset:

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr())
plt.show()
```

## Unemployment Rate Analysis: Data Visualization
Now let’s visualize the data to analyze the unemployment rate. I will first take a look at the estimated number of employees according to different regions of India:

```python
data.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Region",
               "longitude","latitude"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="Region", data=data)
plt.show()
```

Now let’s see the unemployment rate according to different regions of India:

```python
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=data)
plt.show()
```

Now let’s create a dashboard to analyze the unemployment rate of each Indian state by region. For this, I’ll use a sunburst plot:

```python
unemploment = data[["States", "Region", "Estimated Unemployment Rate"]]
figure = px.sunburst(unemploment, path=["Region", "States"], 
                     values="Estimated Unemployment Rate", 
                     width=700, height=700, color_continuous_scale="RdY1Gn", 
                     title="Unemployment Rate in India")
figure.show()
```
